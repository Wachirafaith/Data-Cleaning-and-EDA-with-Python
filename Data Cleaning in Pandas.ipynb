# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
# reading in the file
df = pd.read_excel(r"file:///C:\Users\Admin\Desktop\FAITH\PANDAS%20TUTORIAL\Customer%20Call%20List.xlsx")
df
```

```python
# removing duplicates
df=df.drop_duplicates()
df
```

```python
df
```

```python
# removing columns we dont need
df= df.drop(columns= 'Not_Useful_Column')
df
```

```python
# standardizing the data column by column
```

```python
#df['Last_Name']=df['Last_Name'].str.lstrip("...")
#df['Last_Name']=df['Last_Name'].str.lstrip("/")
#df['Last_Name']=df['Last_Name'].str.rstrip("_")
df['Last_Name']=df['Last_Name'].str.strip("123._/")
df
```

```python
# working with phone numbers
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '_' + x[3:6] + '_' + x[6:10])
df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan__','')
df['Phone_Number']=df['Phone_Number'].str.replace('Na__','')
df
```

```python
# splitting columns
df[['street_address','state','zip_code']]= df["Address"].str.split(',',n=2, expand= True)
df
```

```python
# standardizing column values using replace

df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')

df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')

df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# filling null values
```

```python
df=df.replace('N/a','')
df
```

```python
df=df.fillna('')
df
```

```python
# filtering down rows of data
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
# another way to drop null values
#df.dropna(subset='Phone_Number',inplace = True)
```

```python
df=df.reset_index(drop=True)
df
```
